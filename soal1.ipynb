```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
df = pd.read_csv('survey lung cancer.csv')
df.head()
```

```python
df.columns = ["Gender", "Age", "Smoking", "Yellow_Fingers", "Anxiety", "Peer_Pressure", "Chronic_Disease", "Fatigue", "Allergy", "Wheezing", "Alcohol_Consuming", "Coughing","Shortness_Of_Breath","Swallowing_Difficulty","Chest_Pain","Lung_Cancer"]
```

```python
import matplotlib.pyplot as plt
#plots other feature sets
df.hist(figsize = (15,15))
plt.show()
```

```python
df["Gender"].replace(["M","F"], [0,1], inplace=True)
df["Lung_Cancer"].replace(["NO","YES"], [0,1], inplace=True)
```

```python
df.head()
```

```python
x=df[["Gender","Age","Smoking","Anxiety","Fatigue","Wheezing","Coughing","Shortness_Of_Breath","Chest_Pain"]]

y=df["Lung_Cancer"]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)

print("Number transaction x_train dataset: ", x_train.shape)
print("Number transaction y_train dataset: ", y_train.shape)
print("Number transaction x_test dataset: ", x_test.shape)
print("Number transaction y_test dataset: ", y_test.shape)
```

```python
model = GaussianNB()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
print('Accuracy: ', accuracy_score(y_test, y_pred))
```
