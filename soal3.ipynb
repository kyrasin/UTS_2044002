```python
import pandas as pd
df = pd.read_csv('survey lung cancer.csv')
```

```python
df.columns = ["Gender", "Age", "Smoking", "Yellow_Fingers", "Anxiety", "Peer_Pressure", "Chronic_Disease", "Fatigue", "Allergy", "Wheezing", "Alcohol_Consuming", "Coughing","Shortness_Of_Breath","Swallowing_Difficulty","Chest_Pain","Lung_Cancer"]
```

```python
df["Gender"].replace(["M","F"], [0,1], inplace=True)
df["Lung_Cancer"].replace(["NO","YES"], [0,1], inplace=True)
```

```python
data=pd.get_dummies(df)
data
```

```python
#pisahkan atribute dan label
predictions = ['Gender','Age','Smoking','Anxiety','Fatigue','Wheezing','Coughing','Shortness_Of_Breath','Chest_Pain']

x=data[predictions]
y=data['Lung_Cancer']

#lakukan normalisasi terhadap data yang kita miliki

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(x)
scaled_data = scaler.transform(x)
scaled_data = pd.DataFrame(scaled_data, columns=x.columns)
scaled_data.head()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(scaled_data,y, test_size=0.2, random_state=1)
print(len(y_test))
```

```python
from sklearn import linear_model

model = linear_model.LogisticRegression()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
