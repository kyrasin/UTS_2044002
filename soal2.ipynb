```python
import pandas as pd
from sklearn.datasets import load_iris

iris = pd.read_csv('survey lung cancer.csv')
```

```python
iris.info()
iris.head()
```

```python
iris.columns = ["Gender", "Age", "Smoking", "Yellow_Fingers", "Anxiety", "Peer_Pressure", "Chronic_Disease", "Fatigue", "Allergy", "Wheezing", "Alcohol_Consuming", "Coughing","Shortness_Of_Breath","Swallowing_Difficulty","Chest_Pain","Lung_Cancer"]
```

```python
iris["Gender"].replace(["M","F"], [0,1], inplace=True)
iris["Lung_Cancer"].replace(["NO","YES"], [0,1], inplace=True)
```

```python
x=iris[["Gender","Age","Smoking","Anxiety","Fatigue","Wheezing","Coughing","Shortness_Of_Breath","Chest_Pain"]]

y=iris["Lung_Cancer"]

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.1,random_state=123)
```

```python
from sklearn.tree import DecisionTreeClassifier

tree_model = DecisionTreeClassifier()

tree_model = tree_model.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

y_pred = tree_model.predict(x_test)

acc_secore = round(accuracy_score(y_pred,y_test),3)

print('Accuracy: ',acc_secore)
print('classification report')
print(classification_report(y_pred,y_test))
```
